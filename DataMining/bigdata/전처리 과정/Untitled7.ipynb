```python
import pandas as pd
chiese_all=pd.read_excel('C:/Users/user/Downloads/chinese.xlsx')
pd.options.display.float_format = '{:.2f}'.format
chiese_all
```

```python
del chiese_all["년월"]
del chiese_all["이용자 구분"]
del chiese_all["시도명"]
del chiese_all["데이터기준일자"]

chiese_all
```

```python
sleep=chiese_all[chiese_all["업종"]=="숙박"]
some=chiese_all[chiese_all["업종"] == "소매"]
activity=chiese_all[chiese_all["업종"] == "문화/레져"]
food=chiese_all[chiese_all["업종"] == "식음료"]
shop=chiese_all[chiese_all["업종"]=="쇼핑"]
food
```

```python
some_group=some.groupby('읍면동명',as_index=False).sum()
some_group["1인당이용금액"] = None
some_group["1인당이용금액"]=some_group["이용금액"]/some_group["이용자수"]
some_group["업종"]="소매"
some_group
```

```python
shop_group=shop.groupby('읍면동명',as_index=False).sum()
shop_group["1인당이용금액"] = None
shop_group["1인당이용금액"]=shop_group["이용금액"]/shop_group["이용자수"]
shop_group["업종"]= "쇼핑"
shop_group
```

```python
food_group=food.groupby('읍면동명',as_index=False).sum()
food_group["1인당이용금액"] = None
food_group["1인당이용금액"]=food_group["이용금액"]/food_group["이용자수"]
food_group["업종"]= "음식"
food_group
```

```python
sleep_group=sleep.groupby('읍면동명',as_index=False).sum()
sleep_group["1인당이용금액"] = None
sleep_group["1인당이용금액"]=sleep_group["이용금액"]/sleep_group["이용자수"]
sleep_group["업종"]="숙박"
sleep_group
```

```python
activity_group=activity.groupby('읍면동명',as_index=False).sum()
activity_group["1인당이용금액"] = None
activity_group["1인당이용금액"]=activity_group["이용금액"]/activity_group["이용자수"]
activity_group["업종"]="문화레저"
activity_group
```

```python
chinese=pd.concat([activity_group,food_group,sleep_group,some_group,shop_group])

chinese=chinese.sort_values(by="읍면동명")

chinese
```
