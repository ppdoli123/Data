```python
from tkinter import *
import tkinter as tk
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import matplotlib.pyplot as plt
plt.rc('font', family='NanumGothic')



import pandas as pd
df1 =pd.read_csv('20년도매장수,매출액.csv')
top_8 = df1.groupby('읍면동명').sum().sort_values(by="이용자수",ascending = False).head(8)
top_8
dong = ['노형동','이도2동','연동','송산동','천지동','중앙동','정방동','애월읍']

li = list()

#동마다  업종이용률과 이용금액률 구하기
for key in dong:
    d1 = df1[(df1['읍면동명'])==key]
    peo = d1.groupby('읍면동명').sum()['이용자수'].values[0]
    d1['업종 이용률'] = (d1['이용자수']/peo)
    peo_2 = d1.groupby('읍면동명').sum()['이용금액'].values[0]
    d1['업종별 이용금액률'] = (d1['이용금액'] / peo_2)
    
    result = pd.concat([d1])
    li.append(result)
result1 = pd.concat(li)    

#창업비용 담기

d2 = result1

d4 = d2[(d2['업종명']=='음식')]
d4['창업비용'] = 89000000
d4
d5 = d2[(d2['업종명']=='소매업')]
d5['창업비용'] = 92000000

d6 = d2[(d2['업종명']=='숙박업')]
d6['창업비용'] = 417000000

result = pd.concat([d4,d5,d6])
p = result.sort_values(by = "읍면동명")
p['창업radio'] = p['점포당 매출액']/p['창업비용']
p['점포당 이용률'] = p['이용자수']/p['점포 수']



root = tk.Tk()
rt1= tk.Tk()
rt2= tk.Tk()
rt3= tk.Tk()
rt4= tk.Tk()

    

def openrt2():
    #rt2 디자인r=rt1)
    btn2=PhotoImage(file='bt3.png',master=rt2)
    bg2=PhotoImage(file='png2.png',master=rt2)
    bg_label2=Label(rt2,image=bg2)
    bg_label2.place(x=0,y=0)
    rt2.geometry('300x500+300+0')

    
    rt2_lst=Listbox(rt2,selectmode='single', height=4)

  
    #해당 동만 추출
    for i in ['노형동','이도2동','연동','송산동','천지동','중앙동','정방동','애월읍']:
        
        rt2_lst.insert(END, i)
    
    rt2_lst.place(x=75,y=150)
    
    def getlst2():
        c=rt2_lst.curselection()
        d=rt2_lst.get(c)
        print('선택된 항목:',d)
        #d값을 매개변수로 받음
        def openrt4(d):
            
            #매개변수로 받은 값으로 동 설정
            name = d
            #p에 전체값이 있는데 거기서 해당 동으로 분류
            temp = p[p['읍면동명'].isin([name])]
            #업종 이용률 내림차순 정렬
            #밑에 다 같음
            temp_1 = temp.sort_values('업종 이용률',ascending = False)
            temp_1.reset_index(inplace=True)
            del temp_1['index']
            temp_1.reset_index(inplace=True)
            temp_1

            temp_2 = temp.sort_values('업종별 이용금액률',ascending = False)
            temp_2.reset_index(inplace=True)
            del temp_2['index']
            temp_2.reset_index(inplace=True)

            temp_3 = temp.sort_values('점포당 이용률',ascending = False)
            temp_3.reset_index(inplace=True)
            del temp_3['index']
            temp_3.reset_index(inplace=True)

            temp_4 = temp.sort_values('창업radio',ascending = False)
            temp_4.reset_index(inplace=True)
            del temp_4['index']
            temp_4.reset_index(inplace=True)
            
            #여기서 합치고 동과 업종으로 그루핑함
            result = pd.concat([temp_1,temp_2,temp_3,temp_4])
            tot = result.groupby(['읍면동명','업종명'], as_index=False).sum()
            tot = tot.sort_values(by='index',ascending=True)
            
            #점수및 등수만 남기고 다 날려버림
            tot['점수'] = tot['index']
            del tot['index']
            del tot['점포 수']
            del tot['점포당 매출액']
            del tot['이용금액']
            del tot['창업비용']
            del tot['창업radio']
            del tot['이용자수']
            del tot['업종별 이용금액률']
            del tot['점포당 이용률']
            del tot['업종 이용률']
            tot["등수"]=tot["점수"].rank(method="dense",ascending=True)

            #1등 값 추출 , text에 어떤 업종을 추천하는지 말해줘야 함으로 추출함
            one = tot.loc[tot['등수']==1.0]

            one = one['업종명'].values[0]
            
            #창 크기 변경
            rt4.geometry('1000x700+0+0') #root 창 크기  
            
            #창준이가 해준거 그냥 내 코드만 넣음
            bg=PhotoImage(file='png4.png',master=rt4)
            bg_label=Label(rt4,image=bg)# 배경 레이블 생성
            bg_label.place(x=0,y=0)

            #두번째 레이블 생성
            bg3=PhotoImage(file='png5.png',master=rt4)
            label2=Label(rt4,image=bg3)
            label2.pack(side=LEFT,padx=51)

            #확인 버튼 생성
            cbt=Button(rt4,width=75,height=3,text='확인')
            cbt.pack(side=BOTTOM,pady=10)

            figure=plt.Figure(figsize=(5.2,5), dpi=45) #Figure 객체 생성 및 그래프 크기
            figure1=plt.Figure(figsize=(5.2,5), dpi=45) #Figure 객체 생성 및 그래프 크기
            
            #그래프없앰
            figure2=plt.Figure(figsize=(11,3.5), dpi=45) #Figure 객체 생성 및 그래프 크기
            figure3=plt.Figure(figsize=(11,3.5), dpi=45)
#             figure3=plt.Figure(figsize=(5.2,3), dpi=45) #Figure 객체 생성 및 그래프 크기
#             figure4=plt.Figure(figsize=(5.2,3), dpi=45) #Figure 객체 생성 및 그래프 크기

            ax1= figure.add_subplot(1,1,1) #3-전체 행 , 1-그래프 보여줄 열 , 1-그래프를 보여줄 행 선택
            ax2= figure1.add_subplot(1,1,1)
            ax3= figure2.add_subplot(1,1,1)
            ax4= figure3.add_subplot(1,1,1)
            # ax5= figure4.add_subplot(1,1,1)

            line1=FigureCanvasTkAgg(figure,rt4) 
            line1.get_tk_widget().place(x=479,y=50)

            line2=FigureCanvasTkAgg(figure1,rt4) 
            line2.get_tk_widget().place(x=741,y=50)
            line3=FigureCanvasTkAgg(figure2,rt4) 
            line3.get_tk_widget().pack(side=BOTTOM,pady=5)
            line4=FigureCanvasTkAgg(figure3,rt4) 
            line4.get_tk_widget().pack(side=BOTTOM,pady=5)
            # line5=FigureCanvasTkAgg(figure4,root) 
            # line5.get_tk_widget().place(x=741,y=50)

            #첫번째 레이블 생성
            label1=Label(rt4,width=76,height=2,text=name+'에서 추천하는 업종은 '+one+' 입니다. ')
            label1.place(x=459,y=5)





#             df2= temp_2['업종별 이용금액률'].groupby(temp_2['업종명']).sum()
#             df2.plot(kind='line',ax=ax2, color='r',fontsize='10') # 막대 그래프 형태
            ax2.set_title('업종별 이용금액률')
            
            pie_1 = float(temp_2['업종별 이용금액률'][0])
            pie_2 = float(temp_2['업종별 이용금액률'][1])
            pie_3 = float(temp_2['업종별 이용금액률'][2])
            
            labels2 = '음식', '숙박업', '소매업' 
            pieSizes = [float(pie_1),float(pie_2),float(pie_3)]
            my_colors2 = ['lightblue','lightsteelblue','silver']
            explode2 = (0.1, 0.1, 0.1) 
            ax2.pie(pieSizes, colors=my_colors2, explode=explode2, labels=labels2, autopct='%1.1f%%', shadow=True, startangle=90) ;
    


            df3= temp_3['점포당 이용률'].groupby(temp_3['업종명']).sum()
            df3.plot(kind='barh',ax=ax3, color='b',fontsize='10')
            
            ax3.set_title('점포당 이용률')

            df4= temp_4['창업radio'].groupby(temp_4['업종명']).sum()
            df4.plot(kind='bar',ax=ax4, color='r',fontsize='10')
            ax4.set_title('창업radio')
                
                
            
            ax1.set_title('업종 이용률')
            pie1 = float(temp_1['업종 이용률'][0])
            pie2 = float(temp_1['업종 이용률'][1])
            pie3 = float(temp_1['업종 이용률'][2])

            
            labels2 = '음식', '숙박업', '소매업' 
            pieSizes = [float(pie1),float(pie2),float(pie3)]
            my_colors2 = ['lightblue','lightsteelblue','silver']
            explode2 = (0.1, 0.1, 0.1) 
            ax1.pie(pieSizes, colors=my_colors2, explode=explode2, labels=labels2, autopct='%1.1f%%', shadow=True, startangle=90) ;
    
            plt.show();
            rt4.mainloop();

###
        #얘가 핵심** 버튼을 클릭하면 이 함수가 실행되는데 매개변수를 들고감             
        openrt4(d)
    
    rt2_bt1=Button(rt2,width=150, height = 50,image=btn2,command=getlst2)
    rt2_bt1.place(x=70,y=250)
    
    
    scb=Scrollbar(rt2,width=20,orient='vertical')
    scb.config(command=rt2_lst.yview)
    scb.place(x=210,y=160)
    
    rt2_lst.config(yscrollcommand=scb.set)
    rt2.mainloop()
    
def openrt1():
    #rt1 디자인r=rt1)
    btn1=PhotoImage(file='bt3.png',master=rt1)
    bg1=PhotoImage(file='png3.png',master=rt1)
    bg_label1=Label(rt1,image=bg1)
    bg_label1.place(x=0,y=0)
    rt1.geometry('300x500+300+0')
    
    rt1_lst=Listbox(rt1,selectmode='single', height=0)
    rt1_lst.insert(0,'소매')
    rt1_lst.insert(1,'숙박')
    rt1_lst.insert(2,'식음료')
#     rt1_lst.insert(3,'문화/레져'
    rt1_lst.place(x=75,y=150)
    
    def getlst1():
        a=rt1_lst.curselection()
        b=rt1_lst.get(a)
        print('선택된 항목:',b)

        
        
        
        
        def openrt3(b):
            
            rt3.geometry('300x500+600+0')
            lb1=Label(rt3,width = 20, height = 5,text=b)
            lb1.pack()
        
    
            rt3.mainloop
            
        openrt3(b)
    
    rt1_bt1=Button(rt1,width=150, height = 50,image=btn1,command=getlst1)#,command=closert1
    rt1_bt1.place(x=70,y=250)
    rt1.mainloop()
    

def closert1():
    rt1.destroy()

def closert2():
    rt2.destroy()


    
    
root.geometry('300x500+0+0')
bg=PhotoImage(file='png1.png',master=root)
bg_label=Label(root,image=bg)
bg_label.place(x=0,y=0)

bt_bg1=PhotoImage(file='bt2.png',master=root)
bt1=Button(root,width=150, height=50, image=bt_bg1,command=openrt1)
bt1.place(x=132, y=140)

bt_bg2=PhotoImage(file='bt1.png',master=root)
bt2=Button(root,width=150, height=50, image=bt_bg2,command=openrt2)
bt2.place(x=132, y=215)




root.mainloop()
```
