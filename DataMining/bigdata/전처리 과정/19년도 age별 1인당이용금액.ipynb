```python
import pandas as pd
agedata=pd.read_excel('C:/Users/user/Downloads/연령별데이터.xlsx')
del agedata["시도명"]
del agedata["데이터기준일자"]
del agedata["업종코드"]
del agedata["매장수"]
del agedata["지역구분"]
del agedata["성별"]
del agedata["업종명 대분류"]
del agedata["관광구분"]
del agedata["이용자 구분"]
pd.options.display.float_format = '{:.2f}'.format
agedata=agedata.dropna()

agedata
```

```python
#43 개 동분리
#제주시
agedata = agedata.set_index('읍면동명',inplace=True)
agedata.loc[["이도이동","도남동"],"읍면동명"]="이도2동"
agedata.loc[["일도일동"],"읍면동명"]="일도1동"
agedata.loc[["일도이동"],"읍면동명"]="일도2동"
agedata.loc[["이도일동"],"읍면동명"]="이도1동"
agedata.loc[["삼도일동"],"읍면동명"]="삼도1동"
agedata.loc[["삼도이동"],"읍면동명"]="삼도2동"
agedata.loc[["용담일동"],"읍면동명"]="용담1동"
agedata.loc[["용담이동","용담삼동"],"읍면동명"]="용담2동"
agedata.loc[["화북일동","화북이동"],"읍면동명"]="화북동"
agedata.loc[["삼양일동","삼양이동","삼양삼동","도련일동","도련이동"],"읍면동명"]="삼양동"
agedata.loc[["봉개동","용강동","회천동"],"읍면동명"]="봉개동"
agedata.loc[["아라일동","아라이동","오등동","월평동","영평동"],"읍면동명"]="아라동"
agedata.loc[["해안동"],"읍면동명"]="노형동"
agedata.loc[["외도일동","외도이동","도평동","내도동"],"읍면동명"]="외도동"
agedata.loc[["이호일동","이호이동"],"읍면동명"]="이호동"
agedata.loc[["도두일동","도두이동"],"읍면동명"]="도두동"
#서귀포시
```

```python
#소매업으로 분류
agedata.reset_index(drop=True)
some_group=agedata[(agedata['업종명']=="과실 및 채소 소매업")
    |(agedata["업종명"]=="관광 민예품 및 선물용품 소매업")
   |(agedata["업종명"]=="수산물 소매업")
    |(agedata["업종명"]=="육류 소매업")
    |(agedata["업종명"]=="빵 및 과자류 소매업")
    |(agedata["업종명"]=="화장품 및 방향제 소매업")]
some_group["업종명"]="소매업"

       
some_group_1020=some_group[(some_group['연령대']==10.0)
                           |(some_group["연령대"]==20.0)]
some_group_1020["연령대"] = "청소년"

some_group_3050=some_group[(some_group["연령대"]==30.0)
                          |(some_group["연령대"]==40.0)
                          |(some_group["연령대"]==50.0)]
some_group_3050["연령대"] = "성인"
some_group_6070=some_group[(some_group["연령대"]==60.0)
                          |(some_group["연령대"]==70.0)]
some_group_6070["연령대"] = "노인"
list=[some_group_1020,some_group_3050,some_group_6070]
some=[]
some=[0]*3

for i in range(len(list)):
    age=list[i]
    data=age.groupby(["읍면동명","연령대","업종명"],as_index=False).mean()                        
    data.columns=['읍면동명','연령대',"업종명","1인당 이용금액"]
    some[i]=data
    print(data)

```
