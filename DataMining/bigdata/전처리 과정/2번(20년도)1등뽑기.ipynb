```python
import pandas as pd
import numpy as np
pd.options.display.float_format = '{:,.0f}'.format
food = pd.read_csv('식음매출.csv', index_col = 0)
sleep = pd.read_csv('숙박매출.csv', index_col = 0)
buy = pd.read_csv('소매매출.csv', index_col = 0)
```

```python
#식음료
#매출액
food['창업비용']=89000000
food['매출ratio'] = food['점포당 매출액'] / food['창업비용']
food_1 = food.sort_values('매출ratio',ascending=False)
food_1.head().to_csv('식음료매출액.csv')
food_1.reset_index(inplace=True)
del food_1["index"]
food_1.reset_index(inplace=True)

#점포당이용자수
del food["창업비용"]
del food["매출ratio"]
food['점포당이용자수']=food['이용자수']/food['점포 수']
food_2 = food.sort_values('점포당이용자수',ascending=False)
food_2.head().to_csv('식음료점포당이용자수.csv')
food_2.reset_index(inplace=True)
del food_2["index"]
food_2.reset_index(inplace=True)

#동별이용자수
del food["점포당이용자수"]
food['총이용자수']=food['이용자수'].sum()
food['동별이용자수']=food['이용자수']/food['총이용자수'] * 100
food_3 = food.sort_values('동별이용자수',ascending=False)
food_3.head().to_csv('식음료동별이용자비율.csv')
food_3.reset_index(inplace=True)
del food_3["index"]
food_3.reset_index(inplace=True)

#점수구하기
foodtotal = pd.concat([food_1,food_2,food_3])
foodtotal=foodtotal[["index","읍면동명"]]
foodtotal["점수"]=pd.to_numeric(foodtotal["index"])
del foodtotal["index"]

#등수구하기
food1=foodtotal.groupby(["읍면동명"],as_index=False).sum()
food1["등수"]=food1["점수"].rank(method="dense",ascending=True)
food1=food1.sort_values("등수",ascending=True)
food1
```

```python
#숙박업
#매출액구하기
sleep['창업비용']=417000000
sleep['매출ratio'] = sleep['점포당 매출액'] / sleep['창업비용']
sleep_1 = sleep.sort_values('매출ratio',ascending=False)
sleep_1.head().to_csv('숙박매출액.csv')
sleep_1.reset_index(inplace=True)
del sleep_1["index"]
sleep_1.reset_index(inplace=True)

#점포당이용자수구하기
del sleep["창업비용"]
del sleep["매출ratio"]
sleep['점포당이용자수']=sleep['이용자수']/sleep['점포 수']
sleep_2 = sleep.sort_values('점포당이용자수',ascending=False)
sleep_2.head().to_csv('숙박점포당이용자수.csv')
sleep_2.reset_index(inplace=True)
del sleep_2["index"]
sleep_2.reset_index(inplace=True)


#동별이용자수구하기
del sleep["점포당이용자수"]
sleep['총이용자수']=sleep['이용자수'].sum()
sleep['동별이용자수']=sleep['이용자수']/sleep['총이용자수'] * 100
sleep_3 = sleep.sort_values('동별이용자수',ascending=False)
sleep_3.head().to_csv('숙박동별이용자비율.csv')
sleep_3.reset_index(inplace=True)
del sleep_3["index"]
sleep_3.reset_index(inplace=True)


#점수구하기
sleeptotal = pd.concat([sleep_1,sleep_2,sleep_3])
sleeptotal=sleeptotal[["index","읍면동명"]]
sleeptotal["점수"]=pd.to_numeric(sleeptotal["index"])
del sleeptotal["index"]

#등수구하기
sleep1=sleeptotal.groupby(["읍면동명"],as_index=False).sum()
sleep1.erset_index(inplace=False)
sleep1["등수"]=sleep1["점수"].rank(method="dense",ascending=True)
sleep1=sleep1.sort_values("등수",ascending=True)
sleep1.head()
```
