## **추상요약, 핵심문장 4문장 뽑기, 정답과 오답 뽑기, 핵심키워드 5개 뽑기**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from transformers import BartTokenizer, BartForConditionalGeneration, AutoModel, AutoTokenizer
import numpy as np
from sentence_transformers import SentenceTransformer, util
from typing import List
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')


from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk import pos_tag
import re


model_name = "distilbert-base-nli-mean-tokens"
model = SentenceTransformer(model_name)

# 요약문 생성 함수
def summarize(text):
    model_name = "facebook/bart-large-cnn"
    tokenizer = BartTokenizer.from_pretrained(model_name)
    model = BartForConditionalGeneration.from_pretrained(model_name)

    inputs = tokenizer([text], truncation=True, padding='longest', return_tensors='pt')

    summary_ids = model.generate(inputs['input_ids'], num_beams=4, min_length=200, max_length=220, early_stopping=True)
    summary = [tokenizer.decode(s, skip_special_tokens=True, clean_up_tokenization_spaces=False) for s in summary_ids]

    return summary[0]

#중요문장 추출 함수
def extract_key_sentences(text: str, n_sentences: int) -> list:


    sentences = text.split(". ")
    embeddings = model.encode(sentences)

    similarities = util.pytorch_cos_sim(embeddings, embeddings)

    scores = np.sum(similarities.numpy(), axis=1)
    top_n_indices = np.argsort(-scores)[:n_sentences]
    key_sentences = [sentences[index] for index in top_n_indices]

    return key_sentences


#핵심 단어 뽑는 함수
def preprocess_text(text):
    text = re.sub(r'\W', ' ', text)  # 특수문자를 공백으로 대체
    text = re.sub(r'\s+', ' ', text)  # 연속된 공백 문자를 하나의 공백으로 대체
    text = text.lower().strip()  # 소문자로 변환하고 양쪽 공백 제거
    return text

def remove_stopwords(text, lang="english"):
    stop_words = set(stopwords.words(lang))
    word_tokens = word_tokenize(text)
    filtered_text = [word for word in word_tokens if word not in stop_words]
    return " ".join(filtered_text)

def remove_verbs(text):
    word_tokens = word_tokenize(text)
    pos_tags = pos_tag(word_tokens)  # 품사태깅 수행
    filtered_words = [word for word, pos in pos_tags if pos not in ['VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ']]  # 동사 품사가 아닌 단어만 필터링
    nouns = [word for word, pos in pos_tags if pos in ['NN', 'NNP', 'NNS', 'NNPS']]
    return ' '.join(nouns)


def get_top_n_words(sentence, n=5):
    sentence = preprocess_text(sentence)
    sentence = remove_stopwords(sentence)
    sentence = remove_verbs(sentence)

    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform([sentence])
    feature_array = np.array(vectorizer.get_feature_names_out())
    tfidf_sorting = np.argsort(tfidf_matrix.toarray()).flatten()[::-1]

    top_n_words = feature_array[tfidf_sorting][:n]
    return top_n_words



#유사도비교후 정답과 오답뽑는 함수
def find_most_similar(cor_title: str, titles: list, key_sentences: list, thres: int):

    # 중심문장
    prepared_sentences = key_sentences

    # 임베딩
    prepared_sentence_embeddings = model.encode(prepared_sentences)

    # 유사도 저장 배열
    similarities = []

    for title in titles:
        title_emb = model.encode([title])
        cos_sims = cosine_similarity(prepared_sentence_embeddings, title_emb)
        total_sim = sum(similarity[0] for similarity in cos_sims)
        similarities.append([title, total_sim])

        similarities.sort(key=lambda x: x[1], reverse=True)

    # 제목 및 유사도
    title_sim = []

    # 제목 유사도 저장 및 제외
    for sim in similarities:
        if sim[0] == cor_title:
            title_sim = sim
            similarities.remove(sim)
            break

    # 오답 보기
    wrong_answer = []

    for sim in similarities:
        if len(wrong_answer) == 3:
            break

        if sim[1] > thres:
            wrong_answer.append(sim)

    # 최소값 추가
    wrong_answer.append(similarities[-1])

    return title_sim, wrong_answer

def save_data_to_csv(filename, data):
    data.to_csv(filename, index=False, encoding='utf-8-sig')

#해설문장 뽑는 함수
def solve(answer, wrong1, wrong2, wrong3, wrong4, key_sentences, keywords):

    def print_sentence(choice):
        for key in key_sentences:
            choice = choice + key + '\n'
        choice += "올바르게 풀었는지 한번 더 확인해 보세요.\n"

    def print_word(choice):
        for word in keywords:
            choice += word + '\n'

    # 정답 문장
    answer_solve = "정답입니다! \n해당 문제에 대한 핵심문장입니다.\n"

    print_sentence(answer_solve)

    # 유사한 오답
    sim_wrong_solve = "오답입니다. \n해당 문제에 대한 핵심문장입니다.\n\n"
    print_sentence(sim_wrong_solve)

    sim_wrong_solve += "각 보기별 중심문장과의 유사도를 보여드리겠습니다.\n\n"
    sim_wrong_solve += f"{answer}\n{wrong1}\n{wrong2}\n{wrong3}\n{wrong4}\n\n"
    sim_wrong_solve += "추가적으로 해당 문제의 핵심 키워드를 보여드리겠습니다.\n\n"

    print_word(sim_wrong_solve)

    # 잘못된 오답
    wrong_solve = "오답입니다. \n해당 문제에 대한 핵심문장입니다.\n\n"
    print_sentence(wrong_solve)

    wrong_solve += "각 보기별 중심문장과의 유사도를 보여드리겠습니다.\n\n"
    wrong_solve += f"{answer}\n{wrong1}\n{wrong2}\n{wrong3}\n{wrong4}\n\n"
    wrong_solve += "추가적으로 해당 문제의 핵심 키워드를 보여드리겠습니다.\n\n"

    print_word(wrong_solve)

    wrong_solve += "현재 해당문제에 대해서 추가적인 학습이 필요합니다. 추가 문제를 제공하겠습니다."

    return answer_solve, sim_wrong_solve, wrong_solve

def main(input_file,output_file):
    in_data = pd.read_csv(input_file)

    out_columns = ["Title", "Content", "요약", "핵심 문장", "정답","오답1","오답2","오답3","오답4", "핵심 단어",
                   "정답풀이", "유사오답풀이", "잘못된오답"]
    out_df = pd.DataFrame(columns=out_columns)

    for idx, row in in_data.iterrows():
        title = row["Title"]
        content = row["Content"]
        if len(content) > 220:
            summary = summarize(content)
            key_sentences = extract_key_sentences(summary, 4)

            answer, wrong  = find_most_similar(title, in_data["Title"].tolist(), key_sentences,0.96)
            keywords = get_top_n_words(summary)

            answer_solve, sim_wrong_solve, wrong_solve = solve(answer, wrong[0],
                                                               wrong[1], wrong[2], wrong[3], key_sentences, keywords)

            new_row = {"Title": title, "Content": content, "요약": summary, "핵심 문장": key_sentences,
                    "정답": answer, "오답1": wrong[0], "오답2": wrong[1]
                    ,"오답3": wrong[2], "오답4": wrong[3], "핵심 단어": keywords,
                       "정답풀이": answer_solve, "유사오답풀이": sim_wrong_solve, "잘못된오답": wrong_solve}


            out_df = out_df.append(new_row, ignore_index=True)

    save_data_to_csv(output_file, out_df)


if __name__ == "__main__":
    input_file = "/content/drive/MyDrive/Colab Notebooks/technology.csv"

    output_file = "/content/drive/MyDrive/Colab Notebooks/output_technology.csv"
    main(input_file, output_file)
```

```python
import random
import pandas as pd
from IPython.display import Markdown, display
flag = True
data = pd.read_csv("output_test.csv")
import re

flag = True

def generate(row):
    print('=================================================================================================')
    print("다음글의 주제를 고르시오\n\n")
    print_paragraph(row['요약'])

    options = [
        ( '1','answer',row['정답'][1:row['정답'].rfind(',')]),
        ( '2','good_wrong1',row['오답1'][1:row['오답1'].rfind(',')]),
        ( '3','good_wrong2',row['오답2'][1:row['오답2'].rfind(',')]),
        ( '4','good_wrong3',row['오답3'][1:row['오답3'].rfind(',')]),
        ( '5','bad_wrong',row['오답4'][1:row['오답4'].rfind(',')])
    ]
    answer=[]
    num_index=[]
    random.shuffle(options)
    i =1
    for num,index,option in options:
        print(str(i)+"."+option)
        i += 1
        answer.append(index)
        num_index.append(num)

    input_number = int(input("정답을 고르세요"))

    print('\n\n')

    print('=================================================================================================')
    if answer[input_number-1] == 'answer':
        answer[0]
        print_correct(row)
    if answer[input_number - 1] == 'good_wrong1' or answer[input_number - 1] == 'good_wrong2' or answer[input_number - 1] == 'good_wrong3':
        print_similarAnswer(row,num_index)
    if answer[input_number-1] == 'bad_wrong':
        print_dissimilarAnswer(row,num_index)
        input_number=10
    print('=================================================================================================')

    return input_number,num_index


def print_similarAnswer(row,num_index):
    explaination = row['유사오답풀이'].split('\n\n')
    print_illustration(explaination, row,num_index)

def print_dissimilarAnswer(row,num_index):
    explaination = row['잘못된오답'].split('\n\n')
    print_illustration(explaination, row,num_index)

def print_correct(row):
    print('*정답입니다 ! 해당 지문의 핵심 문장입니다 !*\n')
    print('핵심 문장 :' + row['핵심 문장'])

def print_illustration(explaination, row,num_index):
    similar= []
    print(explaination[0])  # 오답 문장 출력
    print("\n")
    print_paragraph(row['핵심 문장'])
    print('\n\n')  # 핵심 문장 출력
    print(explaination[1] + '\n')  # 보기별 중심문장 유사도 문장 출력
    for i in explaination[2].split(']'):

        similar.append(i)
    for t in num_index:
        index = int(t)
        print(similar[index-1]+"]")
    print("\n")
    print_paragraph("따라서 중심문장과 문장유사도, 핵심단어를 비교한 결과 "+row['Title']+"입니다.\n")
    print(explaination[3] + '\n')  # 핵심 키워드 문장 출력

    print(row['핵심 단어'])  # 핵심 키워드 출력

def print_paragraph(text):
  display(Markdown(text))

while flag:
    answer = []
    random_number = random.randint(0, 788)
    row = data.loc[random_number]
    input_number,num_index = generate(row)
    if input_number == 10:
        print('\n추가적인 문제 풀이가 필요합니다. 추가 문제를 제공하겠습니다.')

        start_Exit = input('진행하겠습니까?[Y/N]')
        print('\n\n')
        if start_Exit == 'Y':
            continue
        elif start_Exit == 'N':
            flag = False
    else:
        flag = False
```
